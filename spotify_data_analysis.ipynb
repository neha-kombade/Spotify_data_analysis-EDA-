```python
top_five_loudest_tracks.plot.barh()
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='loudness',y='song_title',data=top_five_loudest_tracks)

plt.title("TOP 5 LOUDNESS TRAKES")
plt.show()
```

# MULTIPL FEATURE PLOTS

```python
intrest_feature_cols=['tempo','loudness','acousticness','danceability','duration_ms','energy','instrumentalness','liveness','speechiness','valence']
```

```python
for feature_col in intrest_feature_cols:
    pos_data = df[df['target'] == 1][feature_col]
    neg_data = df[df['target'] == 0][feature_col]
    plt.figure(figure=(12,7))
    sns.displot(pos_data,bins=30,label="Positive",color="green")
    sns.displot(neg_data,bins=30,label="Negative",color="red")
    plt.legend(loc="upper right")
    plt.title(f"Positive And Negative Histram plot for {feature_col}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming intrest_feature_cols is a list of column names
for feature_col in intrest_feature_cols:
    pos_data = df[df['target'] == 1][feature_col]
    neg_data = df[df['target'] == 0][feature_col]

    # Create a new figure with subplots
    fig, axes = plt.subplots(figsize=(12, 7), nrows=1, ncols=2)

    # Plot positive data
    sns.histplot(pos_data, bins=30, label="Positive", color="green", ax=axes[0])
    axes[0].legend(loc="upper right")
    axes[0].set_title(f"Positive Histogram plot for {feature_col}")

    # Plot negative data
    sns.histplot(neg_data, bins=30, label="Negative", color="red", ax=axes[1])
    axes[1].legend(loc="upper right")
    axes[1].set_title(f"Negative Histogram plot for {feature_col}")

    # Adjust layout
    plt.tight_layout()

    # Show the plots
    plt.show()
```
